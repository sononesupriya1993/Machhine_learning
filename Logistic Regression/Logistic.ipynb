Logistic Regression



```python
from warnings import filterwarnings
filterwarnings('ignore')
```

Read the training dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

Seprate X and Y(Sepecies)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.value_counts()
```

Create a pipeline for preprocessing of X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

Create a Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

Check model accuracy in train and test

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest, ytest)
```

Predicting the results

```python
ypred_tr = model.predict(xtrain)
ypred_ts = model.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ytrain.head()
```

```python
ypred_ts[0:5]
```

```python
ytest.head()
```

Calculating probabilities

```python
yprob_tr = model.predict_proba(xtrain)
yprob_ts = model.predict_proba(xtest)
```

```python
yprob_tr[0:5]
```

```python
yprob_ts[0:5]
```

```python
yprob_tr[0]
```

```python
model.classes_
```

```python
yprob_ts[0]
```

Evaluating the model


Confusion matrix for test data

```python
model.classes_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf, display_labels=model.classes_)
cfd.plot()
```

Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```

Model gives all precision , recall, accuracy and f1 score beyond 0.80 Hence this is a good model

Predicting out of sample data

```python
xnew = pd.read_csv('iris_sample.csv')
xnew
```

Apply pre.transform

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

Predict the Species

```python
preds = model.predict(xnew_pre)
preds
```

Probabilities

```python
model.classes_
```

```python
probs = model.predict_proba(xnew_pre)
probs
```

Getting max probabilites from above

```python
import numpy as np
max_probs = np.max(probs, axis=1)
max_probs
```

Save the above results in xnew
