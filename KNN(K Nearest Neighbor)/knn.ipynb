## K Nearest Neighbors (KNN)

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

REad The Iris Dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

Species value counts

```python
df['species'].value_counts()
```

```python
df['species'].value_counts().plot(kind='bar')
```

Separate X and Y

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

Apply Train Test and Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python

ytest.shape
```

Create KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=40)
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

Hyperparameter Tuning

```python
params = {'n_neighbors': [3,4,5,6,7,8,9,10,11]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
knn = KNeighborsClassifier()
rscv = RandomizedSearchCV(knn, param_distributions=params, cv=5, scoring='f1_macro')
rscv.fit(xtrain, ytrain)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
best_knn = rscv.best_estimator_
best_knn
```

Evaluate Model on Train and Test

```python
best_knn.score(xtrain, ytrain)
```

```python
best_knn.score(xtest, ytest)
```

Predict the Results for Train and Test

```python
ypred_tr = best_knn.predict(xtrain)
ypred_ts = best_knn.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ypred_ts[0:5]
```

```python
ytrain.head()
```

```python
ytest.head()
```

Plot the confusion Matrix

```python
best_knn.classes_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf, display_labels=best_knn.classes_)
cfd.plot()
```

Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```

Above model has good f1_macro score of 0.98 hence i can use this model for out of sample predictions

Predicting Out Of Sample Data

```python
xnew = pd.read_csv('iris_sample.csv')
xnew.head()
```

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

Get The Prediction along with probability

```python
preds = best_knn.predict(xnew_pre)
preds
```

```python
probs = best_knn.predict_proba(xnew_pre)
probs
```

```python
import numpy as np
max_probs = np.max(probs, axis=1)
max_probs
```

Save Above in xnew

```python
xnew['species_pred']=preds
xnew['prob']=max_probs
```

```python
xnew
```

Save above file to results.csv

```python
xnew.to_csv('KNN_results.csv', index=False)
```
